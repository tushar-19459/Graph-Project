```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/split_column_inflation.csv')

# Plot histogram of column 1
data.plot(kind='hist', bins=10, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Column 1')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
