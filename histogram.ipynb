```python
import pandas as pd
import numpy as np

# Generate 1000 random numbers
data = [np.random.randint(200) for i in range(1000)]
total_rows = 1000
chunk_size = total_rows // 10   # 100 values per column

new_data = {}

for i in range(10):
    start_index = i * chunk_size
    end_index = start_index + chunk_size
    chunk = data[start_index:end_index]
    new_data[f'part_{i+1}'] = chunk

# Convert dictionary to DataFrame
df_parts = pd.DataFrame(new_data)

print(df_parts.head())
# df_parts.to_csv('split_column.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: assuming you already loaded data
# data = pd.read_csv('./splitDataset.csv')

df = {}

# Collect all columns
for i in range(len(data.columns)):
    df[f'col_{i+1}'] = data.iloc[:, i]

col_data = pd.DataFrame(df)

# Flatten all values from all columns into a 1D array
plt.hist(col_data.values.flatten(), bins=10, edgecolor='black')

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of All Columns Combined")
plt.show()
```
